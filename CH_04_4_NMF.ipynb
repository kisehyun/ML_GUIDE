# NMF(Non-Negative Matrix Factorization)
***

NMF는 Truncated SVD 처럼 **행렬 근사** 방식을 사용합니다. 특히 NMF는 원본 행렬의 개별 값이 **모두 양수**인 조건이 충족되면 두 개의 행렬(양수)로 분해 될 수 있다는 기법입니다. 행렬을 분해함으로써 **잠재 요소**를 특성으로 가지게 됩니다. 하나의 원본 행렬을 두 개의 행렬로 분해하게 되면 **길고 가는 행렬(W)**과 **작고 넓은 행렬(H)**로 분해됩니다. W는 잠재 요소의 값이 들어있으며 H는 해당 잠재 요소가 어떻게 구성됐는지 나타냅니다.

```python
from sklearn.decomposition import NMF
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
data = iris.data
```

```python
nmf = NMF(n_components = 2)

nmf_data = nmf.fit_transform(data)
```

```python
plt.scatter(nmf_data[:, 0], nmf_data[:, 1], c = iris.target)
plt.show()
```

NMF는 양수 값만을 가지는 데이터에 훨씬 더 강점을 보이기 때문에 PCA와 SVD와는 다른 결과를 가져다 줍니다. 예를 들어 이미지 픽셀의 경우 양수만 존재하기 때문에 NMF 결과 또한 양수만을 반환해 다른 방식 보다 훨씬 더 자주 사용됩니다.

또한 추천 시스템 영역에서 특히 강점을 보이는데 추천에 사용되는 데이터에서 잠재요소를 추출하여 추천 성능을 끌어올리기도 합니다.
